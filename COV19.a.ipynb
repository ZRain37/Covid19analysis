```python
import pandas as pd
import numpy as np
import matplotlib as mpl
cov19 = pd.read_csv("~/Downloads/country_vaccinations.csv")

cov19.head(65)
```

```python
vacpop = cov19['vaccines'].value_counts()
vacpop.plot.bar(figsize = (20,10), cmap = 'Spectral')
```

```python
cotvac = cov19[['vaccines', 'country']].value_counts()
print(cotvac)
cotvac.plot.bar(figsize = (20,10))
```

```python
now = cov19.loc[:, ['country', 'date', 'people_vaccinated']]
print(now)
```

```python
now_df = cov19.loc[:, ['country', 'people_vaccinated']]
now_group = now_df.groupby('country')
now_max = now_group.max().sort_values('people_vaccinated')
print(now)
now_max.plot.bar(figsize = (20,10))
```
